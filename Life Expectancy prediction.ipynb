```python
num_col= df.select_dtypes(include=['int64','float64'])
for col in num_col:
    for var in df['Country'].unique():
        df.update(df[df.loc[:,'Country']==var][col].replace(np.nan,df[df.loc[:,'Country']==var][col].mean()))
```

```python
df.isnull().sum()
```

```python
sns.displot(df['Population'])
```

```python
df.drop('Population',inplace=True, axis=1)
```

```python
num_col= df.select_dtypes(include=['int64','float64'])
for col in num_col:
    df[col] = df[col].replace(np.nan,df[col].median())
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
df2=df[['Country','Status']]
le= LabelEncoder()
```

```python
df2.loc[:,'Country_enc']=le.fit_transform(df2['Country'])
```

```python
order= {"Developing":0, "Developed":1}
df2.loc[:,'Status_enc']=df2['Status'].map(order)
```

```python
df['Status_enc']=df2['Status_enc']
df['Country_enc']= df2['Country_enc']
```

```python
df.drop('Country',axis=1,inplace= True)
df.drop('Status',axis=1, inplace= True)
```

```python
from sklearn.model_selection import train_test_split
X= df.drop('Life expectancy ',axis=1)
y=df['Life expectancy ']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
```
